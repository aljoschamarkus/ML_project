We can also see, that there are people with a Donor_Age of 0. The variable Donor_Age represents the donors age of the last mail solicitation. Since donors must be at least teenagers, every age below 10 is unrealistic. Therefor we check how many colums with those properties there are and drop them.

```python
# Test: how many people with age less that 10 years old, since this data is unrealistic

df[df["DONOR_AGE"]<10].shape[0]  
```

There are 90 rows with Donor_age  < 10. To not lose any data,we impute those values. With that we exchange the values with the median values of the dataset to prevent the model to learn wrong age information.

```python
# Replace unrealistic values (Values < 10) with nan:

df["DONOR_AGE"] = df["DONOR_AGE"].mask(df["DONOR_AGE"] < 10, np.nan)


# Replacement of nan with median value:
median_age = df["DONOR_AGE"].median()
df["DONOR_AGE"] = df["DONOR_AGE"].fillna(median_age)

```

### Predicting Donation Type

```python
# DONATION_TYPE from TARGET_D 
da = df["TARGET_D"]

conditions = [
    da >= 50,                    # A
    (da >= 20) & (da < 50),      # B
    (da >= 13) & (da < 20),      # C
    (da >= 10) & (da < 13),      # D
    da < 10                      # E
]

choices = ["A", "B", "C", "D", "E"]

df["DONATION_TYPE"] = np.select(conditions, choices)
```

```python
# Features (X) and target (y) definition
data = df.drop(columns=["TARGET_B", "TARGET_D", "DONATION_TYPE"])

y = df["DONATION_TYPE"]

# Train/Test-Split
X_train, X_test, y_train, y_test = train_test_split(
    data,
    y,
    random_state=0,
    stratify=y
)
```

### Tree Classifier:

Preprocessing:

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier


categorical_features = data.select_dtypes(include=["object"]).columns
numeric_features = data.select_dtypes(include=["int64", "float64"]).columns

# One-Hot-Encoding
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
X_train_cat = encoder.fit_transform(X_train[categorical_features])
X_test_cat = encoder.transform(X_test[categorical_features])

# Imputing 
num_imputer = SimpleImputer(strategy="median")
X_train_num = num_imputer.fit_transform(X_train[numeric_features])
X_test_num = num_imputer.transform(X_test[numeric_features])

# Scaling
scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train_num)
X_test_num = scaler.transform(X_test_num)

# Final datasets
X_train_final = np.hstack([X_train_num, X_train_cat])
X_test_final = np.hstack([X_test_num, X_test_cat])

# Featurenames: 
cat_feature_names = encoder.get_feature_names_out(categorical_features)
feature_names = list(numeric_features) + list(cat_feature_names)

print("Number Featurenames:", len(feature_names))
print("Shape X_train_final:", X_train_final.shape)
```
