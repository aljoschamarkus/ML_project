## 0.3. Preprocessing Data

### Handling Missing Values

Since most machine learning classifiers cannot handle missing values, the dataset required preprocessing to address incomplete data.

An initial approach considered **dropping columns with a high proportion of missing values (about 75%)**, followed by removing rows that still contained missing entries:
> **Drop columns:** `pd.DataFrame.dropna(df, axis=1)`  
> **Drop rows:** `pd.DataFrame.dropna(df, axis=0)`

However, this strategy would have resulted in substantial data loss. Therefore, missing values were instead handled using **data imputation** with sklearn’s **`SimpleImputer`**, allowing most instances and features to be retained.

Some categorical variables, such as **`URBANICITY`**, explicitly encode missing information using a placeholder value (`"?"`). According to the data description, `"?"` represents missing urbanicity information rather than a valid category. For consistency with other missing values, these entries were treated as missing and replaced with `np.nan`:
> `df.replace("?", np.nan, inplace=True)`

This preprocessing step ensures a consistent treatment of missing values across the dataset while minimizing unnecessary data loss.

Additionally, we noticed that some entries list a donor age of 0. Since the variable `Donor_Age` reflects the donor’s age at the time of the last mail solicitation, such values are not realistic. Donors must be at least in their teenage years.  
To ensure data quality, we check how many records contain these implausible ages and remove them from the dataset.

```python
# Test: how many people with age less that 10 years old, since this data is unrealistic

df[df["DONOR_AGE"]<10].shape[0]
```

There are 90 rows with a Donor_Age below 10. To avoid losing data, we impute these unrealistic values rather than dropping these.  
By replacing them with the median age of the dataset, we ensure that the model does not learn from incorrect or implausible age information.

```python
# Replace unrealistic values (Values < 10) with nan:

df["DONOR_AGE"] = df["DONOR_AGE"].mask(df["DONOR_AGE"] < 10, np.nan)


# Replacement of nan with median value:
median_age = df["DONOR_AGE"].median()
df["DONOR_AGE"] = df["DONOR_AGE"].fillna(median_age)

```

```python
#Test:
df[df["DONOR_AGE"]<10].shape[0]
df["DONOR_AGE"].describe()

```

After imputation, the minimum donor age is now 12.

Similarly, a value of 0 in `Median_Household_Income` or `Median_Home_Value` is not realistic.  
We address these issues in the same way as before by replacing the implausible zeros with the respective median values of the variables.


```python
# Replace unrealistic values (Values = 0) with nan (MEDIAN_HOME_VALUE):

df["MEDIAN_HOME_VALUE"] = df["MEDIAN_HOME_VALUE"].mask(df["MEDIAN_HOME_VALUE"] == 0, np.nan)


# Replacement of nan with median value:
median_home_value = df["MEDIAN_HOME_VALUE"].median()
df["MEDIAN_HOME_VALUE"] = df["MEDIAN_HOME_VALUE"].fillna(median_home_value)


# Replace unrealistic values (Values = 0) with nan (MEDIAN_HOUSEHOLD_INCOME):

df["MEDIAN_HOUSEHOLD_INCOME"] = df["MEDIAN_HOUSEHOLD_INCOME"].mask(df["MEDIAN_HOUSEHOLD_INCOME"] == 0, np.nan)


# Replacement of nan with median value:
median_income = df["MEDIAN_HOUSEHOLD_INCOME"].median()
df["MEDIAN_HOUSEHOLD_INCOME"] = df["MEDIAN_HOUSEHOLD_INCOME"].fillna(median_income)
```

In the 'URBANICITY' variable, the category "?" appears, which indicates missing or undefined values and must also be imputed.


```python

df[df["URBANICITY"] == '?'].shape[0]
```

We replace the "?" entries with the most frequent category in `URBANICITY`.  
This method is well suited for categorical variables, since measures like the median cannot be computed and the mode provides a reasonable and interpretable substitute.

```python
df = df.replace('?', np.nan)

transformer = Pipeline(
    steps = [
        ("imputer", SimpleImputer(strategy= "most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown = "ignore"))
    ]
)


preprocessor = ColumnTransformer(
    transformers = [
        ("cat", transformer, ["URBANICITY"])
    ]
)
```

### Cross-Validation for Model Evaluation

In accordance with the task requirement, we apply k-fold cross-validation to obtain a more robust estimate of the model’s performance. Instead of relying on a single train–test split, cross-validation evaluates the model across multiple different data splits, resulting in a more stable and reliable assessment of its generalization ability.

We use stratified k-fold cross-validation rather than standard k-fold cross-validation. Stratification ensures that the class distribution is preserved in each fold, which is particularly important in this dataset due to class imbalance. This allows each fold to remain representative of the overall data distribution and leads to fairer and more meaningful evaluation results, especially for minority classes.


```python
#perform stratified 5-fold-cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores = cross_validate(
    log_reg,
    X_train_final,
    y_train,
    cv=cv,
    scoring=["accuracy", "precision", "recall", "f1"]
)

cv_results = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1"],
    "Mean CV Score": [
        cv_scores["test_accuracy"].mean(),
        cv_scores["test_precision"].mean(),
        cv_scores["test_recall"].mean(),
        cv_scores["test_f1"].mean()
    ],
    "Std CV Score": [
        cv_scores["test_accuracy"].std(),
        cv_scores["test_precision"].std(),
        cv_scores["test_recall"].std(),
        cv_scores["test_f1"].std()
    ]
})
print("Logistic Regression (5-fold Stratified CV) results:")
print(cv_results)
```

The cross-validation results indicate that the logistic regression model achieves a relatively high average accuracy of approximately 0.75. However, this value is misleading due to the strong class imbalance in the dataset. The model predominantly predicts the majority class, which inflates the accuracy score.

This behavior is further reflected in the precision, recall, and F1-score. While precision shows moderate variability across folds, the recall is extremely low, indicating that the model fails to correctly identify most positive cases. Consequently, the F1-score remains very low, highlighting poor overall performance on the minority class.

The low standard deviations across folds suggest that the model’s behavior is consistent, but consistently biased toward the majority class. These results emphasize that accuracy alone is not an appropriate evaluation metric for this task and that recall and F1-score are more informative for assessing model performance under class imbalance.


```python
y_cv_pred = cross_val_predict(
    log_reg,
    X_train_final,
    y_train,
    cv=cv,
    method="predict"
)

print("\nClassification report (Cross-Validated Predictions):")
print(classification_report(y_train, y_cv_pred))

print("Confusion matrix (Cross-Validated Predictions):")
cm_cv = confusion_matrix(y_train, y_cv_pred)
print(cm_cv)

ConfusionMatrixDisplay(confusion_matrix=cm_cv).plot()
plt.title("Logistic Regression - Confusion Matrix (CV Predictions)")
plt.show()
```

The cross-validated classification report reveals a strong bias of the logistic regression model toward the majority class (class 0). The model achieves very high recall for class 0, correctly identifying nearly all non-donors, which largely explains the overall accuracy of 0.75.

In contrast, the performance for the minority class (class 1) is very poor. Although the precision is moderate, the recall is extremely low, indicating that only a small fraction of actual positive cases are correctly identified. This behavior is clearly reflected in the confusion matrix, where the number of false negatives vastly outweighs the true positives.

These results confirm that the model struggles to learn meaningful patterns for the minority class under the current setup. While the model is stable across folds, its predictive usefulness is limited for identifying donors.

## Feature Reduction to Improve Model Performance

In this step, we further aim to improve the model performance by removing variables that are either redundant or strongly correlated with other features. Highly correlated variables often encode very similar information and can therefore increase model complexity without providing additional predictive power. This decision is guided by the correlation analysis performed in the previous step, where groups of strongly correlated features were identified.

This feature reduction step is particularly important for logistic regression. As a linear model, logistic regression is sensitive to multicollinearity, which can lead to unstable coefficient estimates, inflated variances, and reduced interpretability. By removing redundant and highly correlated features, we obtain a more stable model with hopefully better generalization performance and more meaningful coefficients.

The selected reduced feature set is therefore expected to improve performance for linear and distance-based models, while also simplifying the overall model structure.


The following variables were removed:

- CONTROL_NUMBER: Unique identifier without predictive value.
- FILE_AVG_GIFT: Redundant feature, identical to LIFETIME_AVG_GIFT_AMT.
- LIFETIME_AVG_GIFT_AMT, LIFETIME_MAX_GIFT_AMT, LIFETIME_MIN_GIFT_AMT, LIFETIME_GIFT_RANGE: Strongly correlated lifetime donation statistics that provide overlapping information. A reduced set of representative lifetime features was retained instead.
- LAST_GIFT_AMT: Highly correlated with recent and lifetime donation averages, contributing limited additional information.
- LIFETIME_CARD_PROM, LIFETIME_PROM: Highly correlated with LIFETIME_GIFT_COUNT, making them redundant indicators of donor engagement.
- MONTHS_SINCE_ORIGIN: Strongly correlated with MONTHS_SINCE_FIRST_GIFT, both describing donor relationship duration.
- PCT_ATTRIBUTE1–4: Highly specific neighborhood-level demographic attributes with limited direct relevance and high redundancy.
- PER_CAPITA_INCOME: Strongly correlated with MEDIAN_HOUSEHOLD_INCOME; the latter was retained as a more interpretable socioeconomic indicator.
- CLUSTER_CODE: High-cardinality geographic identifier that overlaps with other socioeconomic variables and may increase model complexity without improving predictive performance.
- OVERLAY_SOURCE: Indicates data source rather than donor behavior and therefore lacks predictive relevance.

Using Cross-validation for the optimal k 

```python
k_values = range(1, 20)
cv_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train_final, y_train, cv=5, scoring="f1")
    mean_score = scores.mean()
    cv_scores.append(mean_score)
    if k <= 5:
        print(f"{k:2d} | {mean_score:.4f}")

best_k = k_values[np.argmax(cv_scores)]
print("Best k:", best_k)
```

The overall CV-scores are very low, indicating that the model struggles to correctly balance precision and recall.
k = 1 achieves the highest CV-score (0.2620), suggesting that very local decision boundaries capture more positive-class instances.

The fact that k = 1 appears as the best choice in cross-validation may be a consequence of the highly imbalanced class distribution. When the classes are unevenly represented, larger values of k tend to be dominated by the majority class, making it difficult for the model to correctly identify minority-class samples. 

KNN is highly sensitive to class imbalance because its predictions are based on majority voting among the nearest neighbours. When the minority class is underrepresented in the training data, it is much less likely to appear among these neighbours, causing the classifier to almost always predict the majority class. 

To reduce this effect and give the minority class a fair chance of being represented in the local neighbourhood structure, we apply oversampling to balance the class distribution before fitting the KNN model. This ensures that minority samples are more frequently encountered during prediction and improves the model's ability to correctly classify underrepresented cases.

### 2.1.2 Data preprocessing for clustering
Since clustering algorithms, which are normally relying on euclidean distance, are not good at handling categorical data, the data was again analysed. In particular, it was checked that there is no categorical data hidden in columns with a datatype that indicated numerical data (e.g. binary data `type(data)==int`).<p>
This was done by counting unique entries following the logic below:.<br>
>`if (type(column_data) == numerical) & (len(column.unique()) >= threshold): column = numerical`<br>
>`else: column = categorical`

```python
def classify_columns(df, threshold=6):

    numerical = []
    categorical = []

    for col in df.columns:
        col_type = df[col].dtype

        try:
            nunique = df[col].nunique(dropna=True)
        except:
            categorical.append(col)
            continue

        # object columns -> categorical
        if col_type == "object":
            categorical.append(col)
            continue

        # numerical columns
        if pd.api.types.is_numeric_dtype(col_type):
            # binary or few unique values -> categorical
            if nunique <= threshold:
                categorical.append(col)
            # else -> numerical
            else:
                numerical.append(col)
        else:
            categorical.append(col)

    return {
        "numerical": numerical,
        "categorical": categorical
    }

result = classify_columns(df2, threshold=5)

print("Numerical columns:", len(result["numerical"]))
print("Categorical columns:", len(result["categorical"]))
```

As above `np.nan`s need to be handled and again numerical values were imputed using [SimpleImputer](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html). For categorical data `np.nan`s were filled with and `"Unknown"` since no availability might also carry information.<p>
To check if the above worked properly nandrop with df shape feedback was implemented.

#### 2.1.3.1 Hopkins-value
Initially the general structure of the data was examined regarding clusterdness, using the **hopkins-value** as a metric. Generally it compares nearest-neighbor distances of real points to those of randomly generated points in the same space. Hopkins value 1 represents completely clustered, 0.5 meaning random and 0 corresponding to evenly spaced structured.

```python
hopkins_value = hopkins(X_scaled.values, X_scaled.shape[0])
print("hopkins value", hopkins_value)
```

Hopkins values during the progress of the project varied between roughly 0.07 and 0.1. This is relatively low, though it does not represent random data but rather highly structured data organised not in a clustered way but rather in a uniform grid like structure.<p>
This does not give high hopes for finding nicely segregated clusters but rather indicates that they might be heavily overlapping.<p>

### 2.2.2 Evaluating clustering methods
In the following the culstering was evaluated by the metrics we covered in the TP
#### 2.2.2.1 Internal evaluation
Testing the clustering on internal metrics, which are based on the internal geometry such as the "compactness" of the generated clusters, namely:
- **Silhouette Score** measures how well samples fit within their assigned cluster compared to neighboring clusters; values range from −1 to 1, with higher values indicating better separation and values near 0 indicating overlapping clusters.
- **Calinski–Harabasz** Score evaluates the ratio of between-cluster dispersion to within-cluster dispersion; higher values indicate more compact and better-separated clusters.
- **Davies–Bouldin Index** quantifies the average similarity between each cluster and its most similar neighboring cluster; lower values indicate better clustering quality.

```python
for string, clustering_method in clustering_dictionary.items():
    print(f"{string} silhouette_score", metrics.silhouette_score(X_scaled, clustering_method))
    print(f"{string} Calinski Harabasz score", metrics.calinski_harabasz_score(X_scaled, clustering_method))
    print(f"{string} Davies-Bouldin score", metrics.davies_bouldin_score(X_scaled, clustering_method))
    print("\n")
```

Both KMeans and Agglomerative clustering exhibit low silhouette scores, indicating substantial overlap between clusters and suggesting limited intrinsic separability in the donor feature space. KMeans consistently achieves higher silhouette and Calinski–Harabasz scores, along with a slightly lower Davies–Bouldin index, indicating more compact and better-balanced clusters relative to Agglomerative clustering. While absolute internal metric values remain low, relative performance favors KMeans, aligning with the low Hopkins statistic and suggesting weak natural cluster structure.

##### Silhouette plot

Visualizes the distribution of silhouette coefficients for each sample within a cluster, showing how well points fit their assigned cluster relative to neighboring clusters; wide, mostly positive values indicate good separation, while values near zero or negative indicate overlapping or poorly assigned clusters.

```python
def drawSillouette(X, labels, header="" ):
    y_lower =10
    clusters=list(set(labels))
    n_clusters=len(clusters)

    ax1 = plt.gca()
    ax1.set_xlim([-0.5, 1])
    ax1.set_ylim([0, len(X) + (n_clusters) * 3+ y_lower])
    sil_avg = metrics.silhouette_score(X, labels)
    for i,c in enumerate(clusters):
        silhouette_values = metrics.silhouette_samples(X, labels)
        cs_values = silhouette_values[labels == c]
        cs_values.sort()
        size_ci = cs_values.shape[0]
        y_upper = y_lower + size_ci
        color = cm.nipy_spectral(i / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, cs_values, facecolor=color, edgecolor="k", alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_ci, str(c))
        y_lower = y_upper + 3
    ax1.set_title("Silhouette plot "+ header)
    ax1.set_xlabel("Silhouette coefficient")
    ax1.set_ylabel("Clusters")
    ax1.axvline(x=sil_avg, c="r", linestyle="--")

    ax1.set_yticks([])
    plt.show()

drawSillouette(X_scaled, kmeans_clustering.labels_, "kmeans (K=5)")
drawSillouette(X_scaled, hierachical_clustering.labels_, "HCA (complete linkage) K=5")
```

The plots above show similar trends as the scores before already. The clusters of kmeans are more equal in size and contain less negative silhouette sores than agglomerative clustering.<p>
Most importantly this acts as a control and shows that there are 5 distinct clusters and the silhouette score is not falsely inflated by assigning all data to one huge cluster.

#### 2.2.2.2 External evaluation
Testing the clustering on external metrics, based on the correspondence of the generated clusters with the donation amount in `DONATION_TYPE`, namely:
- **Homogeneity Score** measures whether each cluster contains predominantly a single class; values range from 0 to 1, with higher values indicating purer clusters.
- **Completeness Score** measures whether all samples of a given class are assigned to the same cluster; values range from 0 to 1, with higher values indicating better class coverage.
- **V-measure** is the harmonic mean of homogeneity and completeness, balancing cluster purity and class coverage; higher values indicate better overall alignment with external labels.
-  **Fowlkes–Mallows Index (FMI)** measures pairwise agreement between cluster assignments and true labels; values range from 0 to 1, with higher values indicating stronger agreement.
-  **Contingency Matrix** shows the distribution of true labels across clusters and is used to qualitatively assess cluster–label alignment and dominance patterns.

<p>
To get a rough reference on the performance of the clustering it was compared with random label assingement (averaged of 1000 monte carlo samplings).


```python
# Define random sampling
def monte_carlo(method):
    n_runs = 1000
    n_clusters = len(np.unique(kmeans_clustering.labels_))
    random_fmi = []

    for _ in range(n_runs):
        random_labels = np.random.randint(0, n_clusters, size=len(y))
        random_fmi.append(
            method(y, random_labels)
        )
    return np.array(random_fmi).mean()

for string, clustering_method in clustering_dictionary.items():
    print(f"{string} Homogeneity score", metrics.homogeneity_score(y, clustering_method), "random (monte carlo):", monte_carlo(metrics.homogeneity_score))
    print(f"{string} Completeness score", metrics.completeness_score(y, clustering_method), "random (monte carlo):", monte_carlo(metrics.completeness_score))
    print(f"{string} V-measure score", metrics.v_measure_score(y, clustering_method), "random (monte carlo):", monte_carlo(metrics.v_measure_score))
    print(f"{string} FMI score", metrics.fowlkes_mallows_score(y, clustering_method), "random (monte carlo):", monte_carlo(metrics.fowlkes_mallows_score))

    contingency_matrix = cmatrix(clustering_method, y)
    print(f"{string} contingency matrix:\n", contingency_matrix)
```

External evaluation metrics for both methods exceed their respective Monte Carlo random baselines, demonstrating that the obtained clusters capture non-random structure with respect to the external label. However, homogeneity, completeness, and V-measure values remain low, indicating that this structure is weak and does not correspond to well separated label defined groups. KMeans outperforms Agglomerative clustering across all external metrics, including FMI, suggesting slightly stronger alignment with the external label.<p>

Contingency matrix analysis shows substantial label overlap across clusters, with no cluster dominated by a single class. KMeans displays marginally more concentrated label distributions, whereas Agglomerative clustering exhibits more diffuse mixing, consistent with its weaker external performance.

## 2.3. Clustering - Final Discussion and Conclusions

Contrary to initial expectations, clustering donor behavior proved challenging due to substantial overlap between clusters and the absence of strong intrinsic group structure. This difficulty is consistently reflected across multiple analytical perspectives. The Hopkins statistic remained low across different feature subsets, indicating limited natural cluster tendency and suggesting that donor characteristics are better described as a continuum rather than as well-separated groups. PCA visualizations further supported this interpretation by revealing heavy overlap between clusters and gradual transitions in donor characteristics, rather than distinct boundaries.<p>

Internal evaluation metrics, including silhouette, Calinski–Harabasz, and Davies–Bouldin scores, consistently indicated weak geometric separation for all clustering approaches. While KMeans outperformed Agglomerative clustering in relative terms, absolute values remained low, reinforcing the conclusion that no algorithm was able to recover clearly separable clusters. External evaluation metrics showed a similar pattern: homogeneity, completeness, and V-measure scores were modest in magnitude but consistently exceeded Monte Carlo random baselines, demonstrating that the observed clustering captured weak yet non-random structure. The agreement between internal metrics, external metrics, PCA projections, and the Hopkins statistic provides strong evidence that the results are robust and not an artifact of a particular method or evaluation choice.<p>

Despite these limitations, the clustering analysis yielded interpretable and practically meaningful donor profiles. In particular, a subset of clusters aligned well with intuitive behavioral archetypes. For example, one cluster was characterized by younger donors with high donation frequency and relatively small donation amounts, corresponding to young, frequent, low-donation donors. Another cluster captured older donors with lower donation frequency but higher donation amounts, reflecting old, infrequent, high-value donors. These patterns emerged consistently in cluster-level summaries and contingency analyses, even though they did not form sharply separated groups.<p>

Overall, the results suggest that while donor behavior does not naturally partition into discrete classes, unsupervised learning can still provide valuable exploratory segmentation. The clustering outcomes, supported by consistent evidence across multiple evaluation methods, highlight gradual behavioral trends rather than strict categories. This reinforces the view that clustering in this context is best used as an interpretative tool to uncover donor archetypes and behavioral gradients, rather than as a mechanism for precise classification or prediction of donation types.<p>
